A PROJECT ON EMOTION DETECTION WITH OPENCV AND DEEPFACE

```python
import cv2#importing by installing opencv
```

```python
from deepface import DeepFace#importing by first installing deepface
```

```python
img = cv2.imread('happyboy1.jpg')#defining a variable by printing an image by a variable named image.
```

```python
import matplotlib.pyplot as plt#importing matplotlib function from python
```

```python
plt.imshow(img)##by default it will be in bgr(blue green red)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))#converting bgr to rgb means from bgr form to real form of image.
```

```python
predictions = DeepFace.analyze(img)
```

```python
pr=predictions[0]
```

```python
print(predictions[0]['dominant_emotion'])
```

```python
type(predictions[0]['dominant_emotion'])
```

We are trying to draw a rectangle across the face

```python
faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')#harcasscade is basically a face recognition algorithm which is used for reading the face.
```

```python
cv2.CascadeClassifier
```

```python
gray =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#print(facecascade.empty())
faces = faceCascade.detectMultiScale(gray,1.1,4)

#create a rectangular around the faces
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))#converting bgr to rgb means from bgr form to real form of image.
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
#use PutText() method for inserting text in video
cv2.putText(img,
                (predictions[0]['dominant_emotion']),
                (0,50),
                font,1,
                (0,0,255),
                2,
                cv2.LINE_4);
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))#converting bgr to rgb means from bgr form to real form of image.
```
